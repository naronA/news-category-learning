```python
%matplotlib inline
import learning_data as ld
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
news_vecs = ld.load('ldata/2017-10-13.svdtd') 
label = news_vecs[:,0]
data = news_vecs[:,1]
```

```python
mat_dic = {}
for l, d in zip(label,data):
    ind = l.index(1)
    total = None 
    if ind in mat_dic:
        total = mat_dic[ind]
    
    if total is None:
        total = d
    else:
        total += d
    mat_dic.update({ind:total})
```

```python
d_len = len(data[0])
for k,v in mat_dic.items():
    ave = v/d_len
    x = [i for i in range(489,d_len)]
    plt.plot(x, ave[489:], label=k)
plt.legend()
```
